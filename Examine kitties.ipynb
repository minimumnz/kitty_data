```python
import pandas as pd

df = pd.read_csv('feature_combinations.csv', index_col=0)
```

### Find out percent of kitties resulting from breeding with different attributes.

The below is if we have a kitten with the wingtip trait, what are the different combinations of traits that the parents might have had. 

If the parents (feature_a, feature_b) have 'fabulous' and 'raisedbrow' they have 251 offspring, and 39 of them or 15% ended up with wingtips trait.


```python
feature = 'wingtips'
child = (df.child_express == feature)
not_b = (df.feature_b != feature)
not_a = (df.feature_a != feature)
df[ child & not_a & not_b ].sort_values(by='percent_expression', ascending=False)
```
